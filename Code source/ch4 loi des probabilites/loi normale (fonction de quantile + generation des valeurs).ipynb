```python
import scipy.stats as stats
p = [0.25, 0.5, 0.75]   # Liste des probabilités à calculer les quantiles
q = stats.norm.ppf(p)   # Liste des quantiles pour les probabilités dans la liste p
print("Le quantile pour la probabilité p1 est :", q[0])
print("Le quantile pour la probabilité p2 est :", q[1])
print("Le quantile pour la probabilité p3 est :", q[2])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
mu = 2      # Moyenne
sigma = 3   # Écart type
n = 800    # Nombre de valeurs à générer
samples = norm.rvs(mu, sigma, n)

x = np.linspace(-6,6, 100)
plt.hist(samples, bins=27)
plt.xlabel('Valeur')
plt.ylabel('Densité de probabilité')
plt.title('histogramme de 800 valeurs genere ave la loi normale N(2,3)')
plt.show()
```
