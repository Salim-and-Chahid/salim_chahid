```python
from scipy.stats import poisson

lambda_ = 2  # paramètre lambda
p = 0.8     # probabilité

# Calcul de la fonction de quantiles
quantile = poisson.ppf(p, lambda_)
```

```python
print(quantile)
```

```python
import math
import matplotlib.pyplot as plt

lamd = 3
probs = [math.exp(-lamd)]
quantiles = [0]

for i in range(1, 15):
    p = math.exp(-lamd) * lamd**i / math.factorial(i)
    probs.append(p)
    quantiles.append(i)
    
    if sum(probs) >= 0.99:
        break

plt.bar(quantiles, probs)
plt.xlabel('Quantile')
plt.ylabel('Probabilité')
plt.title('Fonction de quantiles pour une loi de Poisson avec λ=3')
plt.show()
```
