```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Paramètres de la loi de Poisson
lambda_ = 2

# Domaine de définition de la fonction de répartition
k = np.arange(0, 20)

# Calcul de la fonction de répartition pour chaque valeur de k
cdf = poisson.cdf(k, lambda_)

# Tracé du graphique de la fonction de répartition
plt.plot(k, cdf, marker='o')

# Ajout de légendes et de titres
plt.xlabel('k')
plt.ylabel('F(k; $\lambda$)')
plt.title('Fonction de répartition de la loi de Poisson')

# Affichage du graphique
plt.show()
```
